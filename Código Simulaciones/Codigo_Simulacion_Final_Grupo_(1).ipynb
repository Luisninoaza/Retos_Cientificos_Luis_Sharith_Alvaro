```python
import numpy as np
import matplotlib.pyplot as plt

# =========================================================
# DATOS TRACKER
# =========================================================

# RECTA 1
t_recta1 = np.array([0.067,0.083,0.100,0.117,0.134,0.150,0.167,0.184,0.200,0.217,0.234,0.250,0.267,0.284,0.300,0.317,0.334])
x_recta1 = np.array([-2.794E-2,-1.728E-2,-8.678E-3,3.615E-3,1.501E-2,2.451E-2,3.517E-2,4.508E-2,5.500E-2,6.656E-2,7.721E-2,8.508E-2,9.573E-2,1.060E-1,1.160E-1,1.270E-1,1.340E-1])
y_recta1 = np.array([7.578E-4,-6.099E-5,3.494E-4,-5.949E-5,-1.042E-3,-5.800E-5,-8.768E-4,-1.532E-3,-2.187E-3,-2.432E-3,-3.250E-3,-2.676E-3,-3.495E-3,-4.150E-3,-4.805E-3,-5.050E-3,-5.049E-3])

# RECTA 2
t_recta2 = np.array([0.050,0.067,0.083,0.100,0.117,0.133,0.150,0.167,0.183,0.200,0.217,0.233,0.250,0.267,0.283,0.300,0.317,0.333,0.350,0.367])
x_recta2 = np.array([-2.299E-2,-1.495E-2,-4.245E-3,4.660E-3,1.629E-2,2.297E-2,3.446E-2,3.992E-2,5.041E-2,5.759E-2,6.700E-2,7.554E-2,8.380E-2,9.321E-2,1.020E-1,1.090E-1,1.180E-1,1.270E-1,1.370E-1,1.420E-1])
y_recta2 = np.array([5.595E-3,5.165E-3,7.392E-3,7.105E-3,6.388E-3,6.173E-3,6.317E-3,5.455E-3,6.318E-3,5.744E-3,5.098E-3,4.309E-3,5.243E-3,4.597E-3,3.807E-3,3.736E-3,1.726E-3,9.364E-4,7.933E-4,-1.433E-3])

# RECTA 3
t_recta3 = np.array([0.067,0.083,0.100,0.117,0.133,0.150,0.167,0.183,0.200,0.217,0.233,0.250,0.267,0.283,0.300,0.317,0.333,0.350,0.367,0.383,0.400,0.417,0.433])
x_recta3 = np.array([-3.547E-2,-2.727E-2,-1.823E-2,-9.461E-3,-9.873E-4,3.535E-3,1.484E-2,2.388E-2,3.377E-2,3.999E-2,4.790E-2,5.638E-2,6.542E-2,7.305E-2,7.983E-2,8.803E-2,9.707E-2,1.050E-1,1.130E-1,1.220E-1,1.280E-1,1.390E-1,1.450E-1])
y_recta3 = np.array([-2.363E-3,-1.111E-4,-4.037E-4,4.390E-3,1.555E-3,2.115E-3,3.516E-3,3.223E-3,4.625E-3,7.161E-3,6.022E-3,6.013E-3,5.720E-3,5.429E-3,6.270E-3,7.108E-3,5.403E-3,6.241E-3,5.102E-3,5.093E-3,5.933E-3,4.791E-3,1.676E-3])

# RECTA 4
t_recta4 = np.array([0.167,0.183,0.200,0.217,0.233,0.250,0.267,0.283,0.300,0.317,0.333,0.350,0.367,0.383,0.400,0.417,0.433,0.450,0.467,0.483])
x_recta4 = np.array([-3.291E-2,-2.290E-2,-1.368E-2,-4.464E-3,6.619E-3,1.567E-2,2.489E-2,3.229E-2,4.321E-2,5.164E-2,5.973E-2,6.851E-2,7.751E-2,8.860E-2,9.621E-2,1.040E-1,1.170E-1,1.260E-1,1.340E-1,1.410E-1])
y_recta4 = np.array([1.139E-4,5.923E-4,5.956E-4,5.990E-4,-1.058E-3,9.898E-4,9.931E-4,1.178E-3,1.566E-3,1.094E-3,8.415E-4,1.510E-4,-1.926E-4,-1.850E-3,-1.318E-3,-7.855E-4,-2.972E-3,-3.187E-3,-4.006E-3,-3.949E-3])

# RECTA 5
t_recta5 = np.array([0.067,0.083,0.100,0.117,0.134,0.150,0.167,0.184,0.200,0.217,0.234,0.250,0.267,0.284,0.300,0.317,0.334,0.351])
x_recta5 = np.array([-1.731E-2,-6.892E-3,1.077E-3,1.506E-2,2.425E-2,3.390E-2,4.426E-2,5.269E-2,6.234E-2,7.225E-2,8.108E-2,9.221E-2,1.030E-1,1.100E-1,1.210E-1,1.290E-1,1.410E-1,1.490E-1])
y_recta5 = np.array([6.359E-3,6.850E-3,5.189E-3,7.375E-3,6.789E-3,7.329E-3,7.053E-3,6.517E-3,7.056E-3,5.656E-3,5.478E-3,5.154E-3,5.237E-3,4.033E-3,4.882E-3,3.579E-3,2.440E-3,2.311E-3])

# RECTA 6
t_recta6 = np.array([0.083,0.100,0.117,0.133,0.150,0.167,0.183,0.200,0.217,0.233,0.250,0.267,0.283,0.300,0.317,0.333,0.350,0.367,0.383])
x_recta6 = np.array([-3.289E-2,-2.213E-2,-1.215E-2,-1.496E-3,7.647E-3,1.883E-2,2.828E-2,3.980E-2,4.936E-2,5.699E-2,6.855E-2,7.694E-2,8.842E-2,9.768E-2,1.080E-1,1.170E-1,1.260E-1,1.350E-1,1.470E-1])
y_recta6 = np.array([-2.302E-3,-3.001E-3,-3.737E-3,-2.056E-3,-2.035E-3,-3.113E-3,-1.090E-3,-9.601E-4,-1.317E-3,-2.956E-3,-3.619E-3,-4.428E-3,-3.504E-3,-5.863E-3,-8.565E-3,-8.544E-3,-8.937E-3,-1.092E-2,-1.317E-2])

# RECTA 7
t_recta7 = np.array([0.150,0.167,0.183,0.200,0.217,0.233,0.250,0.267,0.283,0.300,0.317,0.333,0.350,0.367,0.383,0.400,0.417])
x_recta7 = np.array([-2.682E-2,-1.284E-2,-2.255E-4,1.011E-2,2.281E-2,3.353E-2,4.751E-2,5.832E-2,7.008E-2,7.956E-2,9.127E-2,1.050E-1,1.150E-1,1.300E-1,1.400E-1,1.510E-1,1.590E-1])
y_recta7 = np.array([-3.624E-3,-3.872E-3,-4.504E-3,-5.473E-3,-4.303E-3,-6.645E-3,-6.894E-3,-7.435E-3,-7.121E-3,-7.144E-3,-7.730E-3,-1.021E-2,-1.255E-2,-1.417E-2,-1.244E-2,-1.341E-2,-1.699E-2])

# ROTATIVO 1
t_rot1 = np.array([0.117,0.133,0.150,0.167,0.183,0.200,0.217,0.234,0.250,0.267,0.284,0.300,0.317,0.334,0.350,0.367,0.384,0.400,0.417,0.434])
x_rot1 = np.array([-1.679E-2,-1.152E-2,-9.917E-4,6.938E-3,1.357E-2,2.195E-2,2.641E-2,3.225E-2,3.716E-2,3.736E-2,3.924E-2,4.070E-2,4.135E-2,3.843E-2,3.599E-2,3.313E-2,2.679E-2,2.172E-2,1.489E-2,8.032E-3])
y_rot1 = np.array([-1.057E-3,1.478E-3,6.549E-3,1.253E-2,1.897E-2,2.625E-2,3.271E-2,4.526E-2,5.303E-2,6.609E-2,7.565E-2,8.608E-2,1.000E-1,1.090E-1,1.210E-1,1.330E-1,1.460E-1,1.570E-1,1.670E-1,1.740E-1])

# ROTATIVO 2
t_rot2 = np.array([0.117,0.133,0.150,0.167,0.183,0.200,0.217,0.234,0.250,0.267,0.284,0.300,0.317,0.334,0.350,0.367,0.384,0.400,0.417,0.434,0.450,0.467])
x_rot2 = np.array([-2.071E-2,-1.180E-2,-3.628E-3,8.580E-3,1.602E-2,2.446E-2,3.137E-2,3.907E-2,4.389E-2,4.919E-2,5.579E-2,5.972E-2,6.334E-2,6.727E-2,6.915E-2,6.994E-2,7.167E-2,6.811E-2,6.890E-2,6.696E-2,6.382E-2,6.041E-2])
y_rot2 = np.array([-2.446E-3,-2.447E-3,-1.295E-3,2.423E-3,4.728E-3,9.285E-3,1.368E-2,1.939E-2,2.327E-2,3.150E-2,3.894E-2,4.548E-2,5.507E-2,6.162E-2,7.011E-2,8.032E-2,9.033E-2,1.010E-1,1.110E-1,1.180E-1,1.310E-1,1.400E-1])

# ROTATIVO 3
t_rot3 = np.array([0.083,0.100,0.117,0.133,0.150,0.167,0.184,0.200,0.217,0.234,0.250,0.267,0.284,0.300,0.317,0.334,0.350,0.367,0.384])
x_rot3 = np.array([-3.092E-2,-1.793E-2,-9.053E-3,-3.502E-4,1.103E-2,1.830E-2,2.491E-2,3.209E-2,3.622E-2,4.284E-2,4.784E-2,5.044E-2,5.439E-2,5.242E-2,5.427E-2,5.317E-2,5.149E-2,5.515E-2,6.251E-2])
y_rot3 = np.array([1.482E-3,3.541E-3,5.040E-3,9.977E-3,1.643E-2,2.233E-2,2.784E-2,3.546E-2,4.222E-2,5.394E-2,6.385E-2,7.330E-2,8.350E-2,9.477E-2,1.060E-1,1.200E-1,1.320E-1,1.410E-1,1.460E-1])

# ROTATIVO 4
t_rot4 = np.array([0.133,0.150,0.167,0.184,0.200,0.217,0.234,0.250,0.267,0.284,0.300,0.317,0.334,0.350,0.367,0.384,0.400,0.417,0.434])
x_rot4 = np.array([-2.814E-2,-1.488E-2,-8.234E-3,1.754E-3,7.229E-3,1.712E-2,2.410E-2,3.234E-2,3.878E-2,4.277E-2,4.657E-2,4.900E-2,5.188E-2,5.180E-2,4.944E-2,4.800E-2,4.401E-2,4.104E-2,3.438E-2])
y_rot4 = np.array([3.164E-3,3.810E-3,7.202E-3,1.062E-2,1.436E-2,1.832E-2,2.623E-2,3.324E-2,4.105E-2,5.010E-2,6.024E-2,7.181E-2,8.401E-2,9.403E-2,1.080E-1,1.190E-1,1.320E-1,1.430E-1,1.520E-1])

# ROTATIVO 5
t_rot5 = np.array([0.083,0.100,0.117,0.133,0.150,0.167,0.184,0.200,0.217,0.234,0.250,0.267,0.284,0.300,0.317,0.334,0.350,0.367,0.384,0.400])
x_rot5 = np.array([-2.906E-2,-1.838E-2,-8.630E-3,2.692E-3,1.030E-2,2.069E-2,2.960E-2,3.730E-2,4.333E-2,4.991E-2,5.176E-2,5.797E-2,6.158E-2,6.416E-2,6.322E-2,6.451E-2,6.291E-2,6.113E-2,5.490E-2,5.229E-2])
y_rot5 = np.array([-4.770E-3,-5.315E-3,-4.374E-3,6.532E-4,3.355E-3,9.867E-3,1.545E-2,2.131E-2,2.819E-2,3.748E-2,4.277E-2,5.597E-2,6.340E-2,7.742E-2,8.717E-2,9.832E-2,1.110E-1,1.250E-1,1.370E-1,1.480E-1])

# ROTATIVO 6
t_rot6 = np.array([0.050,0.067,0.083,0.100,0.117,0.133,0.150,0.167,0.183,0.200,0.217,0.234,0.250,0.267,0.284,0.300,0.317,0.334])
x_rot6 = np.array([-1.251E-2,-3.299E-3,6.198E-3,1.512E-2,2.553E-2,3.589E-2,4.331E-2,5.263E-2,6.194E-2,7.097E-2,7.597E-2,8.419E-2,9.138E-2,9.677E-2,1.030E-1,1.060E-1,1.100E-1,1.100E-1])
y_rot6 = np.array([2.594E-2,2.393E-2,2.020E-2,1.992E-2,1.918E-2,2.091E-2,2.109E-2,2.478E-2,2.847E-2,3.389E-2,4.218E-2,5.030E-2,6.037E-2,7.263E-2,8.391E-2,9.611E-2,1.080E-1,1.200E-1])

# ROTATIVO 7
t_rot7 = np.array([0.100,0.117,0.134,0.150,0.167,0.184,0.200,0.217,0.234,0.250,0.267,0.284,0.300,0.317,0.334,0.350,0.367,0.384,0.401,0.417])
x_rot7 = np.array([-2.775E-2,-1.729E-2,-8.771E-3,1.599E-3,9.073E-3,1.711E-2,2.281E-2,3.061E-2,3.478E-2,4.040E-2,4.377E-2,4.674E-2,4.874E-2,4.841E-2,4.904E-2,4.791E-2,4.315E-2,4.314E-2,3.862E-2,4.223E-2])
y_rot7 = np.array([-2.843E-3,3.850E-4,3.932E-3,9.654E-3,1.585E-2,2.246E-2,2.994E-2,3.808E-2,4.532E-2,5.530E-2,6.367E-2,7.260E-2,8.169E-2,9.166E-2,1.010E-1,1.130E-1,1.230E-1,1.350E-1,1.490E-1,1.560E-1])

# =========================================================
# CÍRCULO PLATAFORMA
# =========================================================
angulos = np.linspace(0, 2*np.pi, 300)
circ_x  = 0.15 * np.cos(angulos)
circ_y  = 0.15 * np.sin(angulos)
LIM     = 0.20

# =========================================================
# FUNCIONES FÍSICAS
# =========================================================

def calcular_v0_phi(t, x, y):
    """Velocidad inicial y dirección usando los dos primeros puntos."""
    dx  = x[1] - x[0]
    dy  = y[1] - y[0]
    dt  = t[1] - t[0]
    v0  = np.sqrt(dx**2 + dy**2) / dt
    phi = np.arctan2(dy, dx)
    return v0, phi

def calcular_omega(t, x, y, skip=2):
    """
    Omega promedio con signo correcto:
    calcula dθ/dt sobre los puntos estables (saltando 'skip' iniciales),
    luego elige +/- según cuál produce la curva más cercana a los datos.
    """
    theta      = np.unwrap(np.arctan2(y, x))
    omega_inst = np.diff(theta) / np.diff(t)
    omega_abs  = abs(np.mean(omega_inst[skip:]))

    v0, phi = calcular_v0_phi(t, x, y)
    tau     = t - t[0]

    def dist_media(w):
        xt = x[0] + v0*tau*np.cos(w*tau + phi)
        yt = y[0] + v0*tau*np.sin(w*tau + phi)
        return np.mean(np.sqrt((x-xt)**2 + (y-yt)**2))

    return +omega_abs if dist_media(+omega_abs) <= dist_media(-omega_abs) else -omega_abs


def teo_recta(t, x, y, v0, phi):
    tau   = np.linspace(0, t[-1]-t[0], 300)
    return x[0] + v0*tau*np.cos(phi), y[0] + v0*tau*np.sin(phi)


def teo_rotativa(t, x, y, v0, phi, omega):
    tau   = np.linspace(0, t[-1]-t[0], 300)
    return x[0] + v0*tau*np.cos(omega*tau+phi), y[0] + v0*tau*np.sin(omega*tau+phi)


def distancias_perp(x_exp, y_exp, x_teo, y_teo):
    """
    Para cada punto experimental calcula su distancia mínima a la
    curva teórica (distancia perpendicular real, no solo en x o en y).
    Esto tiene sentido geométrico: mide cuánto se aparta cada punto
    medido de la trayectoria que predice la física.
    """
    dists = []
    for xi, yi in zip(x_exp, y_exp):
        d = np.sqrt((x_teo - xi)**2 + (y_teo - yi)**2)
        dists.append(np.min(d))
    return np.array(dists)


def banda_perpendicular(x_teo, y_teo, sigma):
    """
    Construye los bordes superior e inferior de la banda ±2σ
    desplazando la curva teórica en dirección perpendicular a ella misma.
    Así la banda rodea la curva como un tubo, no como un rectángulo.
    """
    # Vector tangente en cada punto (diferencia entre puntos consecutivos)
    dx = np.gradient(x_teo)
    dy = np.gradient(y_teo)
    norm = np.sqrt(dx**2 + dy**2) + 1e-12
    # Normal perpendicular (rotación 90°)
    nx = -dy / norm
    ny =  dx / norm
    band = 2 * sigma
    x_sup = x_teo + band * nx
    y_sup = y_teo + band * ny
    x_inf = x_teo - band * nx
    y_inf = y_teo - band * ny
    return x_sup, y_sup, x_inf, y_inf


# =========================================================
# GRÁFICA 1 — TRAYECTORIAS (igual que antes)
# =========================================================

def grafica_trayectorias(datos, titulos, titulo_figura):
    N = len(datos)
    fig, axes = plt.subplots(1, N, figsize=(4.2*N, 5))
    if N == 1: axes = [axes]
    fig.suptitle(titulo_figura, fontsize=14, fontweight='bold', y=1.01)

    for ax, (x_exp, y_exp, x_teo, y_teo), titulo in zip(axes, datos, titulos):
        ax.plot(circ_x, circ_y, '--', color='gray', lw=1, alpha=0.5, label='Plataforma')
        ax.axhline(0, color='k', lw=0.5, ls=':')
        ax.axvline(0, color='k', lw=0.5, ls=':')
        ax.plot(x_exp, y_exp, 'o-', lw=1.8, ms=4, color='steelblue', label='Experimental')
        ax.plot(x_teo, y_teo, '--', lw=2,   color='tomato',    label='Teórica')
        ax.scatter(x_exp[0], y_exp[0], s=80, color='black', zorder=6, label='Inicio')
        # flecha de dirección inicial
        dx_a = x_teo[5]-x_teo[0]; dy_a = y_teo[5]-y_teo[0]
        mag  = np.sqrt(dx_a**2+dy_a**2)+1e-12
        ax.annotate('', xy=(x_teo[0]+0.025*dx_a/mag, y_teo[0]+0.025*dy_a/mag),
                    xytext=(x_teo[0], y_teo[0]),
                    arrowprops=dict(arrowstyle='->', color='tomato', lw=2))
        ax.set_xlim(-LIM, LIM); ax.set_ylim(-LIM, LIM)
        ax.set_aspect('equal')
        ax.set_xlabel('x (m)', fontsize=9); ax.set_ylabel('y (m)', fontsize=9)
        ax.set_title(titulo, fontsize=10, fontweight='bold')
        ax.grid(True, alpha=0.3)
        ax.legend(fontsize=7, loc='upper left')

    plt.tight_layout()
    return fig

def grafica_validacion(datos_val, titulos, titulo_figura):
    """
    Para cada trayectoria muestra:
      - La curva teórica (línea central)
      - La banda ±2σ construida perpendicularmente
      - Los puntos experimentales coloreados:
          VERDE  → dentro de la banda (validado)
          ROJO   → fuera de la banda (desviación)
    Además imprime el σ y el % de puntos dentro.
    """
    N = len(datos_val)
    fig, axes = plt.subplots(1, N, figsize=(4.2*N, 5))
    if N == 1: axes = [axes]
    fig.suptitle(titulo_figura, fontsize=14, fontweight='bold', y=1.01)

    print(f'\n  {"Trayectoria":<14} {"σ_perp (cm)":>12}  {"Puntos dentro ±2σ":>18}')
    print('  ' + '─'*50)

    for ax, (x_exp, y_exp, x_teo, y_teo, sigma), titulo in zip(axes, datos_val, titulos):
        # Banda perpendicular
        x_sup, y_sup, x_inf, y_inf = banda_perpendicular(x_teo, y_teo, sigma)

        # Colorear puntos según si están dentro o fuera
        dists   = distancias_perp(x_exp, y_exp, x_teo, y_teo)
        dentro  = dists <= 2*sigma
        pct     = 100 * np.sum(dentro) / len(dentro)

        print(f'  {titulo:<14} {sigma*100:>12.3f}  {np.sum(dentro)}/{len(dentro)} ({pct:.0f}%)')

        # Fondo del eje
        ax.plot(circ_x, circ_y, '--', color='gray', lw=1, alpha=0.4)
        ax.axhline(0, color='k', lw=0.5, ls=':')
        ax.axvline(0, color='k', lw=0.5, ls=':')

        # Banda ±2σ rellena
        ax.fill(
            np.concatenate([x_sup, x_inf[::-1]]),
            np.concatenate([y_sup, y_inf[::-1]]),
            alpha=0.25, color='orange', label=f'Banda ±2σ ({2*sigma*100:.2f} cm)'
        )

        # Curva teórica
        ax.plot(x_teo, y_teo, '-', lw=2, color='tomato', label='Teórica', zorder=3)

        # Puntos dentro (verde) y fuera (rojo)
        ax.scatter(x_exp[dentro],  y_exp[dentro],
                   s=50, color='green',  zorder=5, label=f'Dentro ({np.sum(dentro)})')
        ax.scatter(x_exp[~dentro], y_exp[~dentro],
                   s=50, color='red',    zorder=5, marker='x', lw=2,
                   label=f'Fuera ({np.sum(~dentro)})')

        ax.set_xlim(-LIM, LIM); ax.set_ylim(-LIM, LIM)
        ax.set_aspect('equal')
        ax.set_xlabel('x (m)', fontsize=9); ax.set_ylabel('y (m)', fontsize=9)
        ax.set_title(titulo, fontsize=10, fontweight='bold')
        ax.grid(True, alpha=0.3)
        ax.legend(fontsize=7, loc='upper left')

    plt.tight_layout()
    return fig


# =========================================================
# CONFIGURACIÓN DE TRAYECTORIAS
# =========================================================

trayectorias_rectas = [
    (t_recta1, x_recta1, y_recta1, 'Recta 1'),
    (t_recta2, x_recta2, y_recta2, 'Recta 2'),
    (t_recta3, x_recta3, y_recta3, 'Recta 3'),
    (t_recta4, x_recta4, y_recta4, 'Recta 4'),
    (t_recta5, x_recta5, y_recta5, 'Recta 5'),
    (t_recta6, x_recta6, y_recta6, 'Recta 6'),
    (t_recta7, x_recta7, y_recta7, 'Recta 7'),
]

trayectorias_rotativas = [
    (t_rot1, x_rot1, y_rot1,  8, 'Rotativo 1'),
    (t_rot2, x_rot2, y_rot2,  3, 'Rotativo 2'),
    (t_rot3, x_rot3, y_rot3,  2, 'Rotativo 3'),
    (t_rot4, x_rot4, y_rot4,  2, 'Rotativo 4'),
    (t_rot5, x_rot5, y_rot5,  2, 'Rotativo 5'),
    (t_rot6, x_rot6, y_rot6,  2, 'Rotativo 6'),
    (t_rot7, x_rot7, y_rot7,  2, 'Rotativo 7'),
]


# =========================================================
# PROCESAMIENTO — RECTAS
# =========================================================

datos_tray_r  = []
datos_val_r   = []
titulos_r     = []

for (t, x, y, nombre) in trayectorias_rectas:
    v0, phi      = calcular_v0_phi(t, x, y)
    x_t, y_t     = teo_recta(t, x, y, v0, phi)
    # σ perpendicular: std de las distancias mínimas de cada punto a la curva teórica
    dists        = distancias_perp(x, y, x_t, y_t)
    sigma        = np.std(dists, ddof=1)

    datos_tray_r.append((x, y, x_t, y_t))
    datos_val_r.append((x, y, x_t, y_t, sigma))
    titulos_r.append(nombre)

print('\n' + '='*56)
print('  RECTAS — VALIDACIÓN ±2σ PERPENDICULAR')
print('='*56)

fig1 = grafica_trayectorias(datos_tray_r, titulos_r,
                             'Trayectorias — Marco Inercial (MRU)')
plt.show()

fig2 = grafica_validacion(datos_val_r, titulos_r,
                           'Validación ±2σ — Marco Inercial')
plt.show()


# =========================================================
# PROCESAMIENTO — ROTATIVOS
# =========================================================

datos_tray_rot = []
datos_val_rot  = []
titulos_rot    = []

for (t, x, y, skip, nombre) in trayectorias_rotativas:
    v0, phi      = calcular_v0_phi(t, x, y)
    omega        = calcular_omega(t, x, y, skip=skip)
    x_t, y_t     = teo_rotativa(t, x, y, v0, phi, omega)
    dists        = distancias_perp(x, y, x_t, y_t)
    sigma        = np.std(dists, ddof=1)

    datos_tray_rot.append((x, y, x_t, y_t))
    datos_val_rot.append((x, y, x_t, y_t, sigma))
    titulos_rot.append(nombre)

print('\n' + '='*56)
print('  ROTATIVOS — VALIDACIÓN ±2σ PERPENDICULAR')
print('='*56)

fig3 = grafica_trayectorias(datos_tray_rot, titulos_rot,
                             'Trayectorias — Marco No Inercial (Rotativo)')
plt.show()

fig4 = grafica_validacion(datos_val_rot, titulos_rot,
                           'Validación ±2σ — Marco No Inercial')
plt.show()
```
