```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
Omega = 2.0  # velocidad angular (rad/s)
v0 = 1.0     # velocidad radial (m/s)
t_max = 1.0  # tiempo máximo (s)
dt = 0.05    # paso de tiempo

# Tiempo
t = np.arange(0, t_max, dt)

# Trayectoria en sistema rotante
x_prime = v0 * t * np.cos(Omega * t)
y_prime = v0 * t * np.sin(Omega * t)

# Parámetros del círculo
centro_x = 0.4
centro_y = 0
radio = 0.4

# Generar puntos del círculo
theta_circulo = np.linspace(0, 2*np.pi, 100)
x_circulo = centro_x + radio * np.cos(theta_circulo)
y_circulo = centro_y + radio * np.sin(theta_circulo)

# Crear la gráfica
plt.figure(figsize=(8, 8))
plt.plot(x_prime, y_prime, 'b-', linewidth=2, label='Trayectoria en marco rotante')
plt.scatter(x_prime[::20], y_prime[::20], c=t[::20], cmap='viridis', s=30, zorder=5)
plt.colorbar(label='Tiempo (s)')

# Círculo
plt.plot(x_circulo, y_circulo, 'r-', linewidth=2, label=f'Círculo: (x-2)² + y² = 4')
plt.scatter(centro_x, centro_y, color='red', s=50, marker='o', zorder=10, label='Centro del círculo (2,0)')
# Configuración del gráfico
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid(True, alpha=0.3)
plt.axis('equal')
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.xlabel("x' (m)")
plt.ylabel("y' (m)")
plt.title(f"Trayectoria en sistema rotante\nLanzamiento radial con v₀ = {v0} m/s, Ω = {Omega} rad/s")
plt.legend()
plt.tight_layout()
plt.show()
```
